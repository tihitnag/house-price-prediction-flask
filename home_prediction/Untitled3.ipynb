```python
import pandas as pd
df1['date'] = pd.to_datetime(df1['date'])

df1['year'] = df1['date'].dt.year
df1['month'] = df1['date'].apply(lambda date:date.month)


```

```python
fig=plt.figure(figsize=(15,7))
fig.add_subplot(2, 2, 1)
df1.groupby('month').mean()['price'].plot()
fig.add_subplot(2, 2, 2)
df1.groupby('year').mean()['price'].plot()
```

```python
df1 = df1.drop('date',axis=1)
df1= df1.drop('id',axis=1)
df1 = df1.drop('zipcode',axis=1)
```

```python
df1.head()
```

```python
x=df1.drop('price',axis=1)
y=df1['price'].values
x.head(3)
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import train_test_split

# Split the data into training and testing sets
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=101)
```

```python
#model 1
from sklearn.linear_model import LinearRegression
model = LinearRegression()  
model.fit(x_train, y_train)


```

```python
#evaluate the model (intercept and slope)
print('intercept',model.intercept_)
print('cof',model.coef_)
y_pred=model.predict(x_test)
```

```python
error=y_pred-y_test
error
sns.distplot(error)
```

```python
predition_compasrsion=pd.DataFrame({'actual':y_test,'predict':y_pred})
predition_compasrsion.head(10)
```

```python
# evaluate the performance of the algorithm (MAE - MSE - RMSE)
from sklearn import metrics
print('MAE:', metrics.mean_absolute_error(y_test, y_pred))  
print('MSE:', metrics.mean_squared_error(y_test, y_pred))  
print('RMSE:', np.sqrt(metrics.mean_squared_error(y_test, y_pred)))
print('VarScore:',metrics.explained_variance_score(y_test,y_pred))
```
